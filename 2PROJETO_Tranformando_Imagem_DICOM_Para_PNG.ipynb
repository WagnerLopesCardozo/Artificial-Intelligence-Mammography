#<font color=hotpink>Introduo

##<font color=hotblue>O Sistema DICOM (Digital Imaging and Communications in Medicine)  um padro amplamente utilizado na rea mdica para aquisio, armazenamento, transmisso e visualizao de imagens mdicas. Ele foi desenvolvido para permitir a troca de informaes e imagens entre diferentes sistemas de imagem e equipamentos mdicos, independentemente do fabricante.

##<font color=hotblue>O Sistema DICOM abrange uma variedade de modalidades de imagem, como radiografia geral, mamografia, tomografia computadorizada (CT), ressonncia magntica (MRI), ultrassom e muitas outras. Ele define um formato de arquivo comum para armazenar dados de imagem mdica, alm de especificar mtodos de comunicao para transmitir esses dados entre sistemas.

##<font color=hotblue>O padro DICOM no apenas inclui as prprias imagens mdicas, mas tambm informaes adicionais importantes, como dados do paciente, informaes sobre o exame, resultados de medies e anotaes. Isso permite que as imagens sejam associadas ao contexto clnico correto e facilitem a interpretao para o laudo pelos profissionais de sade.

##<font color=hotblue>Com o DICOM,  possvel garantir a interoperabilidade entre diferentes dispositivos mdicos e sistemas de sade, facilitando o compartilhamento de informaes e imagens mdicas entre hospitais, clnicas e profissionais de sade. Ele desempenha um papel fundamental no armazenamento e gerenciamento de imagens mdicas digitais, contribuindo para um diagnstico mais preciso, gil e um atendimento ao paciente mais eficiente.

#<font color=hotpink>Na mamografia:

##<font color=hotblue>Nos exames de mamografia, o sistema DICOM,  amplamente utilizado na rotina cotidiana feitos em equipamentos denominados de mamgrafos. A mamografia  uma modalidade de imagem diagnstica que utiliza radiao ionizante (Raios X) para obter imagens detalhadas das mamas, com o objetivo de detectar precocemente possveis anormalidades na estrutura anatmica da mama, como massas tumorais, calcificaes ou outro achado patolgico.

##<font color=hotblue>Com este padro,  possvel realizar o armazenamento e transmisso das imagens de mamografia, permitindo que elas sejam compartilhadas e visualizadas em sistemas de imagem mdica compatveis. Com o uso do DICOM, as imagens de mamografia podem ser associadas s informaes relevantes do paciente, como dados clnicos, resultados de exames anteriores e outras informaes pertinentes.

##<font color=hotblue>Alm disso, o DICOM tambm possibilita a padronizao dos formatos de arquivo e das informaes contidas nas imagens de mamografia, o que facilita a interpretao e comparao das imagens por Radiologistas com especialidade em mamografia e outros profissionais de sade como Mdicos Mastologistas. Isso contribui para um diagnstico mais preciso e eficiente, alm de possibilitar a integrao das imagens de mamografia com outros sistemas de informao mdica, como PACS (Picture Archiving and Communication System) e RIS (Radiology Information System).

##<font color=hotblue>Portanto, o sistema DICOM desempenha um papel importante na mamografia, permitindo a troca de informaes e imagens entre diferentes dispositivos de mamografia, clnicas e hospitais, e facilitando a anlise e interpretao das imagens para auxiliar no diagnstico mdico.

#<font color=hotpink>Algoritmo 1
##<font color=hotblue>Possibilita fazer o uploud do arquivo contendo a imagem para a Computao em Nuvem

```python
import pydicom
from PIL import Image
from google.colab import files

# Carregar arquivo DICOM
uploaded = files.upload()

# ler arquivo DICOM
dicom_data = pydicom.dcmread(next(iter(uploaded)))

# Obtendo dados de pixel
pixel_array = dicom_data.pixel_array

# Convertendo imagem
image = Image.fromarray(pixel_array)

# Salvando imagem
image.save("output.png")

# Baixar o arquivo PNG
files.download("output.png")
```

#<font color=hotpink>Algoritmo 4
##<font color=hotblue>Consegue carregar e tranformar vrios arquivos e exibilos corretamente, mas com perda de informao aps transformao

```python
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

# Carrega arquivo DICOM
uploaded_files = files.upload()

# Reune os arquivos carregados
for filename in uploaded_files.keys():
    try:
        # Ler arquivo DICOM
        dicom_data = pydicom.dcmread(filename)
    except pydicom.errors.InvalidDicomError:
        print(f"Skipping invalid DICOM file: {filename}")
        continue

    # Obtem dados de pixel
    pixel_array = dicom_data.pixel_array

    # Converter para imagem PIL
    image = Image.fromarray(pixel_array)

    # Cria figura em subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # Exibe a imagem original DICOM
    axes[0].imshow(pixel_array, cmap=plt.cm.gray)
    axes[0].axis('off')
    axes[0].set_title('DICOM Image')

    # Exibe a imagem convertida PNG
    axes[1].imshow(image, cmap=plt.cm.gray)
    axes[1].axis('off')
    axes[1].set_title('PNG Image')

    # Ajusta o espaamento entre os subplots
    plt.tight_layout()

    # Exibe as imagens na tela
    plt.show()
```

#<font color=hotpink>Algoritmo 5
##<font color=hotblue>Para resolver o problema da perda de informao ocorrida nos algoritmos 2, 3 e 4, foi preservado a tonalidade original da escala de cinza ao salvar a imagem DICOM como PNG, usando o mapa de cores da biblioteca e funo matplotlib.cm.gray, diretamente ao exibir e salvar a imagem.

##<font color=hotblue>Ento, a imagem DICOM original  exibida usando o mapa de cores plt.cm.gray, garantindo que a tonalidade da escala de cinza seja preservada. Ao salvar a imagem DICOM como PNG, a funo plt.imsave()  usada com os parmetros vmin e vmax definidos para os valores mnimo e mximo de pixel da imagem DICOM, para manter a mesma faixa de tons de cinza. A imagem PNG convertida  exibida usando o mapa de cores padro, resultando em uma imagem muito semelhante  imagem DICOM original.

```python
import pydicom
import matplotlib.pyplot as plt
from google.colab import files

# Carrega o arquivo DICOM
uploaded_files = files.upload()

# Reune os arquivos carregados
for filename in uploaded_files.keys():
    # Ler arquivo DICOM
    dicom_data = pydicom.dcmread(filename)

    # Obter dados de pixel
    pixel_array = dicom_data.pixel_array

    # Cria uma figura com subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # Exibe a imagem original DICOM
    axes[0].imshow(pixel_array, cmap=plt.cm.gray, vmin=pixel_array.min(), vmax=pixel_array.max())
    axes[0].axis('off')
    axes[0].set_title('DICOM Image')

    # Salva a imagem DICOM como PNG sem alterar a tonalidade da escala de cinza
    png_filename = filename.replace('.dcm', '.png')
    plt.imsave(png_filename, pixel_array, cmap=plt.cm.gray, vmin=pixel_array.min(), vmax=pixel_array.max())

    # Exibe a imagem convertida PNG
    png_image = plt.imread(png_filename)
    axes[1].imshow(png_image, cmap=plt.cm.gray)
    axes[1].axis('off')
    axes[1].set_title('PNG Image')

    # Ajusta o espaamento entre os subplots
    plt.tight_layout()

    # Exibe as imagens na tela
    plt.show()
```

#<font color=hotpink>Algoritmo 6

##<font color=hotblue>Neste algoritmo, o arquivo DICOM  carregado para a Computao em Nuvem pelo usurio, a imagem DICOM  normalizada para ter valores de pixel entre 0 e 1. Em seguida, ele cria uma figura com duas sub-figuras, plota a imagem DICOM na primeira sub-figura e salva temporariamente a imagem DICOM como um arquivo PNG. A imagem PNG  lida e plotada na segunda sub-figura. O arquivo PNG temporrio  removido no final.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# L o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtm a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image = dicom_image / np.max(dicom_image)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Salva a imagem DICOM temporariamente como PNG
temp_dicom_path = 'temp_dicom.png'
plt.imsave(temp_dicom_path, dicom_image, cmap='gray')

# L a imagem PNG temporria
png_image = plt.imread(temp_dicom_path)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(png_image, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove o arquivo PNG temporrio
os.remove(temp_dicom_path)

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>algoritmo 7
##<font color=hotblue>Nessa verso atualizada do algoritmo, aps normalizar a imagem DICOM entre 0 e 1,  feita a multiplicao por 255, para obter uma escala de 0 a 255. Em seguida,  convertido a matriz de pixels para o tipo uint8. Isso permite um melhor ajuste de contraste e brilho na exibio da imagem DICOM.

##<font color=hotblue>Ao plotar a imagem DICOM,  definido vmin e vmax como os valores mnimo e mximo da imagem DICOM para garantir que todos os detalhes sejam visveis. Para a imagem PNG,  mantido o padro de escala entre 0 e 1.

##<font color=hotblue>Dessa forma, as imagens exibidas devem preservar as caractersticas de tonalidade de cinza da imagem DICOM, o que no  observado nas imagens, logo o algoritmo tem que ser melhorado, para mostrar os detalhes da estrutura anatmica interna da mama com mais clareza.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# L o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtm a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image = dicom_image / np.max(dicom_image)

# Ajusta o contraste e brilho da imagem DICOM
dicom_image = dicom_image * 255
dicom_image = dicom_image.astype(np.uint8)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray', vmin=np.min(dicom_image), vmax=np.max(dicom_image))
axs[0].set_title('Imagem de mamografia DICOM')

# Salva a imagem DICOM temporariamente como PNG
temp_dicom_path = 'temp_dicom.png'
plt.imsave(temp_dicom_path, dicom_image, cmap='gray', vmin=np.min(dicom_image), vmax=np.max(dicom_image))

# L a imagem PNG temporria
png_image = plt.imread(temp_dicom_path)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(png_image, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove o arquivo PNG temporrio
os.remove(temp_dicom_path)

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 8
##<font color=hotblue>Nesta verso do algoritmo, a imagem DICOM  normalizada entre 0 e 1 usando a frmula (valor_pixel - valor_mnimo) / (valor_mximo - valor_mnimo). Em seguida, a imagem DICOM normalizada  exibida na primeira sub-figura.

##<font color=hotblue>Para a segunda sub-figura, a imagem DICOM normalizada  convertida em uma imagem PNG e os limites de intensidade so ajustados para corresponder aos valores mnimos e mximos da imagem DICOM. Isso preserva todas as caractersticas de tonalidade de cinza da imagem DICOM na imagem PNG exibida.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# L o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtm a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image_normalized = (dicom_image - np.min(dicom_image)) / (np.max(dicom_image) - np.min(dicom_image))

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image_normalized, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image_normalized, cmap='gray')
png_image.set_clim(np.min(dicom_image_normalized), np.max(dicom_image_normalized))

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image_normalized, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 9
##<font color=hotblue>Nesta nova verso do algoritmo, alm de normalizar a imagem DICOM entre 0 e 1,  aplicado a tcnica de janelamento para melhorar a visualizao. O centro e a largura da janela so obtidos a partir do arquivo DICOM e usados para determinar os valores mnimos e mximos para a imagem DICOM.

##<font color=hotblue>Em seguida,  usado a funo np.clip para limitar os valores da imagem DICOM dentro da faixa definida pela janela. Depois,  feito a normalizao novamente a imagem DICOM dentro do intervalo entre 0 e 1.

##<font color=hotblue>Essas alteraes melhoram ainda mais a visualizao da imagem DICOM, destacando os detalhes da estrutura anatmica e preservando as caractersticas de tonalidade de cinza ao converter para a imagem PNG.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# L o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtm a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Aplica a tcnica de janelamento para melhorar a visualizao
window_center = dicom_file.WindowCenter
window_width = dicom_file.WindowWidth

dicom_image_min = window_center - (window_width / 2)
dicom_image_max = window_center + (window_width / 2)

dicom_image = np.clip(dicom_image, dicom_image_min, dicom_image_max)
dicom_image = (dicom_image - dicom_image_min) / (dicom_image_max - dicom_image_min)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray', vmin=0, vmax=1)
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image, cmap='gray', vmin=0, vmax=1)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image, cmap='gray', vmin=0, vmax=1)
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 10
##<font color=hotblue>Nesta outra verso mais atualizada do algoritmo, foi aplicado a equalizao de histograma na imagem DICOM, usando a funo np.histogram. Foi calculando ento o histograma da imagem DICOM e em seguida, feito a funo de distribuio cumulativa (CDF) normalizada. Em seguida, foi usado a funo np.interp, para mapear os valores da imagem DICOM original para os valores normalizados do CDF. Isso resulta na imagem DICOM equalizada.

##<font color=hotblue>Aps a equalizao de histograma,  feito o plot das imagens DICOM e PNG nas duas sub-figuras, e ento finalmente podemos ter o comparativo das duas imagens no formato DICOM e PNG e ter uma boa percepo dos detalhes da estrutura anatmica da mama.





```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# L o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtm a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Equaliza o histograma da imagem DICOM
hist, bins = np.histogram(dicom_image.flatten(), bins=256, range=(np.min(dicom_image), np.max(dicom_image)))
cdf = hist.cumsum()
cdf_normalized = cdf / cdf[-1]
dicom_image_eq = np.interp(dicom_image.flatten(), bins[:-1], cdf_normalized).reshape(dicom_image.shape)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image_eq, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image_eq, cmap='gray')

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image_eq, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```
