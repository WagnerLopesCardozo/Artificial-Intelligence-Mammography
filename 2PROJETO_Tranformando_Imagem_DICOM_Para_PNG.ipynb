#<font color=hotpink>Introdução

##<font color=hotblue>O Sistema DICOM (Digital Imaging and Communications in Medicine) é um padrão amplamente utilizado na área médica para aquisição, armazenamento, transmissão e visualização de imagens médicas. Ele foi desenvolvido para permitir a troca de informações e imagens entre diferentes sistemas de imagem e equipamentos médicos, independentemente do fabricante.

##<font color=hotblue>O Sistema DICOM abrange uma variedade de modalidades de imagem, como radiografia geral, mamografia, tomografia computadorizada (CT), ressonância magnética (MRI), ultrassom e muitas outras. Ele define um formato de arquivo comum para armazenar dados de imagem médica, além de especificar métodos de comunicação para transmitir esses dados entre sistemas.

##<font color=hotblue>O padrão DICOM não apenas inclui as próprias imagens médicas, mas também informações adicionais importantes, como dados do paciente, informações sobre o exame, resultados de medições e anotações. Isso permite que as imagens sejam associadas ao contexto clínico correto e facilitem a interpretação para o laudo pelos profissionais de saúde.

##<font color=hotblue>Com o DICOM, é possível garantir a interoperabilidade entre diferentes dispositivos médicos e sistemas de saúde, facilitando o compartilhamento de informações e imagens médicas entre hospitais, clínicas e profissionais de saúde. Ele desempenha um papel fundamental no armazenamento e gerenciamento de imagens médicas digitais, contribuindo para um diagnóstico mais preciso, ágil e um atendimento ao paciente mais eficiente.

#<font color=hotpink>Na mamografia:

##<font color=hotblue>Nos exames de mamografia, o sistema DICOM, é amplamente utilizado na rotina cotidiana feitos em equipamentos denominados de mamógrafos. A mamografia é uma modalidade de imagem diagnóstica que utiliza radiação ionizante (Raios X) para obter imagens detalhadas das mamas, com o objetivo de detectar precocemente possíveis anormalidades na estrutura anatômica da mama, como massas tumorais, calcificações ou outro achado patológico.

##<font color=hotblue>Com este padrão, é possível realizar o armazenamento e transmissão das imagens de mamografia, permitindo que elas sejam compartilhadas e visualizadas em sistemas de imagem médica compatíveis. Com o uso do DICOM, as imagens de mamografia podem ser associadas às informações relevantes do paciente, como dados clínicos, resultados de exames anteriores e outras informações pertinentes.

##<font color=hotblue>Além disso, o DICOM também possibilita a padronização dos formatos de arquivo e das informações contidas nas imagens de mamografia, o que facilita a interpretação e comparação das imagens por Radiologistas com especialidade em mamografia e outros profissionais de saúde como Médicos Mastologistas. Isso contribui para um diagnóstico mais preciso e eficiente, além de possibilitar a integração das imagens de mamografia com outros sistemas de informação médica, como PACS (Picture Archiving and Communication System) e RIS (Radiology Information System).

##<font color=hotblue>Portanto, o sistema DICOM desempenha um papel importante na mamografia, permitindo a troca de informações e imagens entre diferentes dispositivos de mamografia, clínicas e hospitais, e facilitando a análise e interpretação das imagens para auxiliar no diagnóstico médico.

#<font color=hotpink>Algoritmo 1
##<font color=hotblue>Possibilita fazer o uploud do arquivo contendo a imagem para a Computação em Nuvem

```python
import pydicom
from PIL import Image
from google.colab import files

# Carregar arquivo DICOM
uploaded = files.upload()

# ler arquivo DICOM
dicom_data = pydicom.dcmread(next(iter(uploaded)))

# Obtendo dados de pixel
pixel_array = dicom_data.pixel_array

# Convertendo imagem
image = Image.fromarray(pixel_array)

# Salvando imagem
image.save("output.png")

# Baixar o arquivo PNG
files.download("output.png")
```

#<font color=hotpink>Algoritmo 3

```python
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

# Carrega arquivo DICOM
uploaded = files.upload()

# Ler arquivo DICOM
dicom_data = pydicom.dcmread(next(iter(uploaded)))

# Obter dados de pixel
pixel_array = dicom_data.pixel_array

# Converte para imagem PIL
image = Image.fromarray(pixel_array)

# Cria uma figura com subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Exibe imagem DICOM original
axes[0].imshow(pixel_array, cmap=plt.cm.gray)
axes[0].axis('off')
axes[0].set_title('DICOM Image')

# Exibe imagem PNG convertida
axes[1].imshow(image, cmap=plt.cm.gray)
axes[1].axis('off')
axes[1].set_title('PNG Image')

# Ajusta o espaçamento entre os subplots
plt.tight_layout()

# Exibe as imagens na tela
plt.show()
```

#<font color=hotpink>Algoritmo 4
##<font color=hotblue>Consegue carregar e tranformar vários arquivos e exibilos corretamente, mas com perda de informação após transformação

```python
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

# Carrega arquivo DICOM
uploaded_files = files.upload()

# Reune os arquivos carregados
for filename in uploaded_files.keys():
    try:
        # Ler arquivo DICOM
        dicom_data = pydicom.dcmread(filename)
    except pydicom.errors.InvalidDicomError:
        print(f"Skipping invalid DICOM file: {filename}")
        continue

    # Obtem dados de pixel
    pixel_array = dicom_data.pixel_array

    # Converter para imagem PIL
    image = Image.fromarray(pixel_array)

    # Cria figura em subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # Exibe a imagem original DICOM
    axes[0].imshow(pixel_array, cmap=plt.cm.gray)
    axes[0].axis('off')
    axes[0].set_title('DICOM Image')

    # Exibe a imagem convertida PNG
    axes[1].imshow(image, cmap=plt.cm.gray)
    axes[1].axis('off')
    axes[1].set_title('PNG Image')

    # Ajusta o espaçamento entre os subplots
    plt.tight_layout()

    # Exibe as imagens na tela
    plt.show()
```

#<font color=hotpink>Algoritmo 5
##<font color=hotblue>Para resolver o problema da perda de informação ocorrida nos algoritmos 2, 3 e 4, foi preservado a tonalidade original da escala de cinza ao salvar a imagem DICOM como PNG, usando o mapa de cores da biblioteca e função matplotlib.cm.gray, diretamente ao exibir e salvar a imagem.

##<font color=hotblue>Então, a imagem DICOM original é exibida usando o mapa de cores plt.cm.gray, garantindo que a tonalidade da escala de cinza seja preservada. Ao salvar a imagem DICOM como PNG, a função plt.imsave() é usada com os parâmetros vmin e vmax definidos para os valores mínimo e máximo de pixel da imagem DICOM, para manter a mesma faixa de tons de cinza. A imagem PNG convertida é exibida usando o mapa de cores padrão, resultando em uma imagem muito semelhante à imagem DICOM original.

```python
import pydicom
import matplotlib.pyplot as plt
from google.colab import files

# Carrega o arquivo DICOM
uploaded_files = files.upload()

# Reune os arquivos carregados
for filename in uploaded_files.keys():
    # Ler arquivo DICOM
    dicom_data = pydicom.dcmread(filename)

    # Obter dados de pixel
    pixel_array = dicom_data.pixel_array

    # Cria uma figura com subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # Exibe a imagem original DICOM
    axes[0].imshow(pixel_array, cmap=plt.cm.gray, vmin=pixel_array.min(), vmax=pixel_array.max())
    axes[0].axis('off')
    axes[0].set_title('DICOM Image')

    # Salva a imagem DICOM como PNG sem alterar a tonalidade da escala de cinza
    png_filename = filename.replace('.dcm', '.png')
    plt.imsave(png_filename, pixel_array, cmap=plt.cm.gray, vmin=pixel_array.min(), vmax=pixel_array.max())

    # Exibe a imagem convertida PNG
    png_image = plt.imread(png_filename)
    axes[1].imshow(png_image, cmap=plt.cm.gray)
    axes[1].axis('off')
    axes[1].set_title('PNG Image')

    # Ajusta o espaçamento entre os subplots
    plt.tight_layout()

    # Exibe as imagens na tela
    plt.show()
```

#<font color=hotpink>Algoritmo 6

##<font color=hotblue>Neste algoritmo, o arquivo DICOM é carregado para a Computação em Nuvem pelo usuário, a imagem DICOM é normalizada para ter valores de pixel entre 0 e 1. Em seguida, ele cria uma figura com duas sub-figuras, plota a imagem DICOM na primeira sub-figura e salva temporariamente a imagem DICOM como um arquivo PNG. A imagem PNG é lida e plotada na segunda sub-figura. O arquivo PNG temporário é removido no final.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# Lê o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtém a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image = dicom_image / np.max(dicom_image)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Salva a imagem DICOM temporariamente como PNG
temp_dicom_path = 'temp_dicom.png'
plt.imsave(temp_dicom_path, dicom_image, cmap='gray')

# Lê a imagem PNG temporária
png_image = plt.imread(temp_dicom_path)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(png_image, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove o arquivo PNG temporário
os.remove(temp_dicom_path)

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>algoritmo 7
##<font color=hotblue>Nessa versão atualizada do algoritmo, após normalizar a imagem DICOM entre 0 e 1, é feita a multiplicação por 255, para obter uma escala de 0 a 255. Em seguida, é convertido a matriz de pixels para o tipo uint8. Isso permite um melhor ajuste de contraste e brilho na exibição da imagem DICOM.

##<font color=hotblue>Ao plotar a imagem DICOM, é definido vmin e vmax como os valores mínimo e máximo da imagem DICOM para garantir que todos os detalhes sejam visíveis. Para a imagem PNG, é mantido o padrão de escala entre 0 e 1.

##<font color=hotblue>Dessa forma, as imagens exibidas devem preservar as características de tonalidade de cinza da imagem DICOM, o que não é observado nas imagens, logo o algoritmo tem que ser melhorado, para mostrar os detalhes da estrutura anatômica interna da mama com mais clareza.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# Lê o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtém a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image = dicom_image / np.max(dicom_image)

# Ajusta o contraste e brilho da imagem DICOM
dicom_image = dicom_image * 255
dicom_image = dicom_image.astype(np.uint8)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray', vmin=np.min(dicom_image), vmax=np.max(dicom_image))
axs[0].set_title('Imagem de mamografia DICOM')

# Salva a imagem DICOM temporariamente como PNG
temp_dicom_path = 'temp_dicom.png'
plt.imsave(temp_dicom_path, dicom_image, cmap='gray', vmin=np.min(dicom_image), vmax=np.max(dicom_image))

# Lê a imagem PNG temporária
png_image = plt.imread(temp_dicom_path)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(png_image, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove o arquivo PNG temporário
os.remove(temp_dicom_path)

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 8
##<font color=hotblue>Nesta versão do algoritmo, a imagem DICOM é normalizada entre 0 e 1 usando a fórmula (valor_pixel - valor_mínimo) / (valor_máximo - valor_mínimo). Em seguida, a imagem DICOM normalizada é exibida na primeira sub-figura.

##<font color=hotblue>Para a segunda sub-figura, a imagem DICOM normalizada é convertida em uma imagem PNG e os limites de intensidade são ajustados para corresponder aos valores mínimos e máximos da imagem DICOM. Isso preserva todas as características de tonalidade de cinza da imagem DICOM na imagem PNG exibida.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# Lê o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtém a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Normaliza os valores de pixel entre 0 e 1
dicom_image_normalized = (dicom_image - np.min(dicom_image)) / (np.max(dicom_image) - np.min(dicom_image))

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image_normalized, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image_normalized, cmap='gray')
png_image.set_clim(np.min(dicom_image_normalized), np.max(dicom_image_normalized))

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image_normalized, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 9
##<font color=hotblue>Nesta nova versão do algoritmo, além de normalizar a imagem DICOM entre 0 e 1, é aplicado a técnica de janelamento para melhorar a visualização. O centro e a largura da janela são obtidos a partir do arquivo DICOM e usados para determinar os valores mínimos e máximos para a imagem DICOM.

##<font color=hotblue>Em seguida, é usado a função np.clip para limitar os valores da imagem DICOM dentro da faixa definida pela janela. Depois, é feito a normalização novamente a imagem DICOM dentro do intervalo entre 0 e 1.

##<font color=hotblue>Essas alterações melhoram ainda mais a visualização da imagem DICOM, destacando os detalhes da estrutura anatômica e preservando as características de tonalidade de cinza ao converter para a imagem PNG.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# Lê o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtém a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Aplica a técnica de janelamento para melhorar a visualização
window_center = dicom_file.WindowCenter
window_width = dicom_file.WindowWidth

dicom_image_min = window_center - (window_width / 2)
dicom_image_max = window_center + (window_width / 2)

dicom_image = np.clip(dicom_image, dicom_image_min, dicom_image_max)
dicom_image = (dicom_image - dicom_image_min) / (dicom_image_max - dicom_image_min)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image, cmap='gray', vmin=0, vmax=1)
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image, cmap='gray', vmin=0, vmax=1)

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image, cmap='gray', vmin=0, vmax=1)
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```

#<font color=hotpink>Algoritmo 10
##<font color=hotblue>Nesta outra versão mais atualizada do algoritmo, foi aplicado a equalização de histograma na imagem DICOM, usando a função np.histogram. Foi calculando então o histograma da imagem DICOM e em seguida, feito a função de distribuição cumulativa (CDF) normalizada. Em seguida, foi usado a função np.interp, para mapear os valores da imagem DICOM original para os valores normalizados do CDF. Isso resulta na imagem DICOM equalizada.

##<font color=hotblue>Após a equalização de histograma, é feito o plot das imagens DICOM e PNG nas duas sub-figuras, e então finalmente podemos ter o comparativo das duas imagens no formato DICOM e PNG e ter uma boa percepção dos detalhes da estrutura anatômica da mama.





```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pydicom
from google.colab import files

# Solicita o upload do arquivo DICOM
uploaded = files.upload()

# Lê o arquivo DICOM
dicom_path = next(iter(uploaded))
dicom_file = pydicom.dcmread(dicom_path)

# Obtém a imagem DICOM em formato de matriz
dicom_image = dicom_file.pixel_array

# Equaliza o histograma da imagem DICOM
hist, bins = np.histogram(dicom_image.flatten(), bins=256, range=(np.min(dicom_image), np.max(dicom_image)))
cdf = hist.cumsum()
cdf_normalized = cdf / cdf[-1]
dicom_image_eq = np.interp(dicom_image.flatten(), bins[:-1], cdf_normalized).reshape(dicom_image.shape)

# Cria um objeto figura com duas sub-figuras
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plota a imagem DICOM na primeira sub-figura
axs[0].imshow(dicom_image_eq, cmap='gray')
axs[0].set_title('Imagem de mamografia DICOM')

# Converte a imagem DICOM em formato PNG
png_image = plt.imshow(dicom_image_eq, cmap='gray')

# Plota a imagem PNG na segunda sub-figura
axs[1].imshow(dicom_image_eq, cmap='gray')
axs[1].set_title('Imagem de mamografia PNG')

# Remove os eixos das sub-figuras
for ax in axs:
    ax.axis('off')

# Exibe a figura
plt.show()
```
